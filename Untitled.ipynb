```python
import pandas as pd
df = pd.read_csv('dados/dados_agregados_limpos.csv')
```

```python
df['Status'].unique()
```

```python
# ADICIONANDO RODADA 0

AtletaIDs = df['AtletaID'].unique() # seleciona atletas

naomodificado= ['AtletaID', 'ClubeID', 'Participou', 'Posicao','Preco', 'ano', 'Apelido', 'Status', 'mes', 'dia', 'variable'] #preco modifica

new_data = pd.DataFrame()
# construindo nova tabela
for ano in df['ano'].unique()[:]:
    for atletaID in AtletaIDs[:]:
        # seleciona os dados da primeira rodada
        data = df[(df['AtletaID'] == atletaID) & (df['Rodada'] == 1) & (df['ano'] == ano)]# Tipo SELECT tabela WHERE
        if data.shape[0] != 0:
            # fazendo atualizações para rodada 0
            for keyn in df.keys():
                if keyn in naomodificado:
                    if(keyn == 'Preco'):# calcula o valor do preço de inicio
                        data['Preco'] = data['Preco'] - data['PrecoVariacao']
                else:
                    data[keyn] = 0.0 # 
            new_data = pd.concat([new_data, data])
df = pd.concat([df, new_data])
```

```python
df
```

```python
# vamos adicionas as variveis de target agora
# dado um atleta id, rodada, clube id, queremos adicionar um campo referente a variação(pontos ou precovar) da rodada t+1
dic = {'Target_Pontos':[], 'Target_varPreco':[]}

for i in range(df.shape[0]): # para toda tabela
    id_ = df.iloc[i]['AtletaID']
    rodada = df.iloc[i]['Rodada']
    clubeid = df.iloc[i]['ClubeID']
    if rodada < 38:
        try: # Se o jogador tiver jogado na proxima partida
            varPreco = df[(df['AtletaID'] == id_ ) & (df['Rodada'] == rodada+1) & (df['ClubeID'] == clubeid)]['PrecoVariacao'].values[0]
            Ponto = df[(df['AtletaID'] == id_ ) & (df['Rodada'] == rodada+1) & (df['ClubeID'] == clubeid)]['Pontos'].values[0]
        except: # Se não tiver jogado, zera os pontos
            varPreco = 0.0
            Ponto = 0.0
    else:
        varPreco = 0.0
        Ponto = 0.0

    dic['Target_Pontos'].append(Ponto)
    dic['Target_varPreco'].append(varPreco)
df['Target_varPreco'] = dic['Target_varPreco']
df['Target_Pontos'] = dic['Target_Pontos']
```

```python
# normalizando
from sklearn.preprocessing import OneHotEncoder
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

def normaliza_XY(X,Y): 
    x = X.select_dtypes(include = ['bool', 'float64', 'int64']).values #returns a numpy array
    min_max_scaler = StandardScaler()#MinMaxScaler() #StandardScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    
    min_max_scaler2 = StandardScaler()#MinMaxScaler()#StandardScaler
    y_ = min_max_scaler2.fit_transform(Y.reshape(-1, 1))

    only_categoric = X.select_dtypes(include=['object'])
    enc = OneHotEncoder()
    enc.fit(only_categoric)
    x_cat = enc.transform(only_categoric).toarray()
    X_normalizado = np.concatenate((x_cat, x_scaled), axis = 1)
    
    
    return X_normalizado, y_.reshape(1, -1)[0], enc, min_max_scaler2, min_max_scaler


def normaliza2_XY(X,Y, min_max_scaler,  min_max_scaler2, enc): 
    x = X.select_dtypes(include = ['bool', 'float64', 'int64']).values #returns a numpy array
    #min_max_scaler = StandardScaler()#MinMaxScaler() #StandardScaler()
    x_scaled = min_max_scaler.transform(x)
    
    #min_max_scaler2 = StandardScaler()#MinMaxScaler()#StandardScaler
    y_ = min_max_scaler2.transform(Y.reshape(-1, 1))

    only_categoric = X.select_dtypes(include=['object'])
    #enc = OneHotEncoder()
    #enc.fit(only_categoric)
    x_cat = enc.transform(only_categoric).toarray()
    X_normalizado = np.concatenate((x_cat, x_scaled), axis = 1)
    
    
    return X_normalizado, y_.reshape(1, -1)[0]
```

```python
def get_best_time(df, limiar, rodada):
    df_rodada = df[df['Rodada'] == rodada]
    values = df_rodada['Prev_Pontos'].values
    sort_values = sorted(values)
    print(sort_values)
    sum_pontos = 0.0
    sum_var = 0.0
    gasto = 0
       
    
    dic_control = {'zag':0, 'lat': 0, 'mei': 0, 'ata': 0, 'gol': 0}
    dic_lim = {'zag':2, 'lat': 2, 'mei': 4, 'ata': 2, 'gol': 1}
    for i in range(len(sort_values)):
        bests = df[df['Prev_Pontos'] == sort_values[len(sort_values)-i -1]]
        for j in range(bests.shape[0]):
            pos = bests.iloc[j]['Posicao']
            if dic_control[pos] < dic_lim[pos] and gasto < limiar:
                dic_control[pos] += 1
                sum_pontos += bests.iloc[j]['Target_Pontos']
                sum_var += bests.iloc[j]['Target_varPreco']
                gasto += bests.iloc[j]['Preco']
                
            if dic_control[pos] == dic_lim[pos]:
                return sum_pontos, sum_var, gasto, dic_control
            
    return sum_pontos, sum_var, gasto, dic_control    
```
